```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Problem 1 (Sinusoid Plot):

def sinusoid(x_data, A, w):
    y = A * np.sin(x_data * w)
    return y
```

```python
x_data = np.linspace(0, 2*np.pi , 1000)
A = np.array([2, 4, 7, 8, 9])
w = np.array([1, 3, 5, 6, 10])
```

```python
plt.figure()
for i, element in enumerate(A):
    y_data = sinusoid(x_data, element, w[i])
    plt.plot(x_data,y_data)
plt.title("5 Different Sine Functions")
plt.xlabel("x data")
plt.ylabel("y data")
plt.legend(['sine function 1', 'sine function 2','sine function 3','sine function 4','sine function 5'], loc="lower left")
plt.show()
```

```python
# Problem 2 (Randomness Plot):

num_list1 = [random.randint(0, 100) for i in range(40)]
num_list2 = [random.randint(0, 100) for i in range(40)]
num_list3 = [random.randint(0, 100) for i in range(40)]
```

```python
plt.figure(figsize= (8,8))
plt.plot(num_list1, num_list2, color = 'orange', linewidth = 10)
plt.plot(num_list1, num_list3, color = 'red', linestyle = 'dashed')
plt.title("randomness")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(['x', 'y'], loc="upper right")
plt.show()
```
